```python
import pandas as pd
import matplotlib.pyplot as plt
import altair as alt
```

```python
df = pd.read_csv("../data/US Wage Data 2021.csv")
df.head()
```

```python
for column in df.columns:
    print(df[column].unique())
```

```python
earnest_1 = alt.Chart(df).mark_circle().encode(
    x=alt.X("Race and ethnicity", axis=alt.Axis(labelAngle=0)),
    y=alt.Y("Median Earnings (USD)", aggregate="mean"),
    color=alt.Color("Sex"),
    size=alt.value(200),
).properties(
    width=500,
    title=alt.Title(
        ["Men have higher median earnings than women", "across Race and Ethnicity groups"],
        fontSize=24,
        align='left',
        anchor='start',
        subtitle="This data has been averaged across Occupation Groups.",
        subtitleFontSize=18,
    ),
)
earnest_1.save("earnest_1.png")
earnest_1
```

```python
deceptive_1 = alt.Chart(df).mark_bar().encode(
    x=alt.X("Race and ethnicity", axis=alt.Axis(labelAngle=0)),
    y=alt.Y("Median Earnings (USD)", aggregate="mean", stack=None, scale=alt.Scale(domain=[40000,75000], zero=False)), #, axis=alt.Axis(labelLimit=)),
    color=alt.Color("Sex"),
    # size=alt.value(200),
).properties(
    width=500,
    title=alt.Title(
        ["Men have higher median earnings than women", "across Race and Ethnicity groups"],
        fontSize=24,
        align='left',
        anchor='start',
        subtitle="This data has been averaged across Occupation Groups.",
        subtitleFontSize=18,
    ),
)
deceptive_1.save("deceptive_1.png")
deceptive_1
```
