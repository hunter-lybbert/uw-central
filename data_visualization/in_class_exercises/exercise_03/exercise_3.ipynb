```python
import pandas as pd
import numpy as np
import altair as alt
```

```python
df = pd.read_json("https://vega.github.io/vega-datasets/data/udistrict.json")
df.head()
```

```python
alt.Chart(df).mark_tick(height=50).encode(
    x=alt.X("lat:Q", scale=alt.Scale(domain=[df["lat"].min() - 1e-3, df["lat"].max() + 1e-3])),
).properties(
    width=600,
    height=100,
)
```

```python
len(df.key.unique())
```

```python
alt.Chart(df).transform_density(
    'lat',
    as_=['lat', 'density'],
    groupby=['key'],
    bandwidth=1e-3,
    extent=[df["lat"].min() - 1e-2, df["lat"].max() + 1e-2],
).mark_area().encode(
    x="lat:Q",
    y='density:Q',
).properties(
    width=200,
    height=100,
).facet(
    'key:N',
    columns=4
)
```

```python
alt.Chart(df).transform_density(
    'lat',
    as_=['lat', 'density'],
    groupby=['key'],
    bandwidth=1e-3,
    extent=[df["lat"].min() - 1e-2, df["lat"].max() + 1e-2],
).mark_area().encode(
    x="lat:Q",
    y=alt.Y('density:Q'),
    color='key:N',
    opacity=alt.value(.5),
).properties(
    width=400,
    height=300,
)
```

```python
alt.Chart(df).transform_density(
    'lat',
    as_=['lat', 'density'],
    groupby=['key'],
    bandwidth=1e-3,
    extent=[df["lat"].min() - 1e-2, df["lat"].max() + 1e-2],
).mark_area().encode(
    x="lat:Q",
    y=alt.Y('density:Q', stack=None),
    color='key:N',
    opacity=alt.value(.5),
).properties(
    width=100,
    height=50,
)
```

```python
alt.Chart(df).transform_density(
    'lat',
    as_=['lat', 'density'],
    groupby=['key'],
    bandwidth=1e-3,
    extent=[df["lat"].min() - 1e-2, df["lat"].max() + 1e-2],
).mark_area().encode(
    x=alt.X(alt.repeat("column"), type="quantitative"),
    y=alt.Y(alt.repeat("row"), type="quantitative"),
    color='key:N',
    opacity=alt.value(.5),
).properties(
    width=100,
    height=50,
).repeat(
    row=list(df.key.unique()),
    column=list(df.key.unique()),
)
```
